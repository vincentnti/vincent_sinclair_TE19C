```
import numpy as np
import matplotlib.pyplot as plt

#3.
def f(x):
  return 3*x+1

array1 = f(np.arange(0,11))

plt.plot(array1)
plt.title('Graph for 3*x+1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```
import numpy as np
import matplotlib.pyplot as plt

#4.
def f(x):
  return x**2-1

array1 = f(np.arange(-2,3))
print(array1)

plt.plot(array1)
plt.title('Graph for x^2-1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
