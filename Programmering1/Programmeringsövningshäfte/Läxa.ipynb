```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sysselsattning.csv', header=1, encoding = 'ISO-8859-1') 

# 1.
# (a)
arbetande = df[df.sysselsättning == 'förvärvsarbetande']
arbetande = arbetande[arbetande.ålder == '40-44 år']
arbetande_total = arbetande['2019'].sum()
print('Total förvärvsarbetande:', arbetande_total)

# (b)
ejarbetande = df[df.sysselsättning == 'ej förvärvsarbetande']
ejarbetande = ejarbetande[ejarbetande.ålder == '40-44 år']
ejarbetande_total = ejarbetande['2019'].sum()
print('Total ej förvärvsarbetande:', ejarbetande_total)

# (c)

labels = 'Arbetande', 'Ejarbetande'
sizes = [arbetande_total, ejarbetande_total]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels)
ax1.axis('equal')
plt.show()
```

```python
# 2.

# (a)
män_arbetar = df[df.kön == 'män']
män_arbetar = män_arbetar[män_arbetar.sysselsättning == 'förvärvsarbetande']
total_män = män_arbetar['2019'].sum()
print('Totalt Arbetande Män från ålder 16-44:', total_män)

# (b)
kvinnor_arbetar = df[df.kön == 'kvinnor']
kvinnor_arbetar = kvinnor_arbetar[kvinnor_arbetar.sysselsättning == 'förvärvsarbetande']
total_kvinnor = kvinnor_arbetar['2019'].sum()
print('Totalt Arbetande Kvinnor från ålder 16-44:', total_kvinnor)

# (c) och (d) Samma fråga för två uppgifter men antar att du vill att jag ska ha kvinnor på (d)
män_arbetar_inte = df[df.kön == 'män']
män_arbetar_inte = män_arbetar_inte[män_arbetar_inte.sysselsättning == 'ej förvärvsarbetande']
total_män_inte = män_arbetar_inte['2019'].sum()
print('Totalt Inte Arbetande Män från ålder 16-44:', total_män_inte)

kvinnor_arbetar_inte = df[df.kön == 'kvinnor']
kvinnor_arbetar_inte = kvinnor_arbetar_inte[kvinnor_arbetar_inte.sysselsättning == 'ej förvärvsarbetande']
total_kvinnor_inte = kvinnor_arbetar_inte['2019'].sum()
print('Totalt Inte Arbetande Kvinnor från ålder 16-44:', total_män_inte)

# (e)
fig = plt.figure()
ax = fig.add_axes([0,1,2,3])
sizes = [total_män, total_män_inte, total_kvinnor, total_kvinnor_inte]
labels = ['Män job', 'Män inte job', 'Kvinnor job', 'Kvinnor inte job']
ax.bar(labels, sizes)
plt.show()

```
